```python
def insertion_sort(items):
    """ Insertion sort """
    items = items.copy()
    swaps = 0
    for i, _ in enumerate(items):
        k = i
        while (k > 0) and (items[k] < items[k - 1]):
            items[k], items[k - 1] = items[k - 1], items[k]
            k -= 1
            swaps += 1
    return items, swaps

A = [6, 10, 4, 5, 1, 2]
A = [-11308, 70868, -37914, 87041, -42905, 17146, 61513, -29073, 6518, 10848, 78985, 42681, 35863, -61067, 4587, 43022, -28444, 71443, -45823, 22646, -68412, -18204, 96516, 41028, -32210, 85911, 36664, 59865, -72837, -25551, -38742, 13228, -90594, 88135, -67028, -71272, 66927, 50927, 64726, -68774, 72003, 47752, -57132, 34729, 31686, 22572, 1817, 96901, -81291, -3760, 54472, -50039, 65748, -34590, -26184, 37315, 91106, -54387, -27138, 30692, 25181, 24952, -30314, 4853, 31412, -15121, -79680, 26198, 70575, 39023, 26410, -70617, 88643, 46177, 207, 83246, 40389, -11610, 32771, 79569, 47547, -98813, 33426, 13087, -22202, -37636, 60316, 16639, 54690, -82523, -30633, -25779, -97295, 33674, -74160, -73875, 5445, -12522, -91223, -42219, -34487, 96519, -2853, -72078, -71957, 84406, 80712, 70973, -83467, 65883, 854, 45497, -52580, 63279, -27737, 14765, -48058, 13760, 96608, -79377, -82346, -37883, -83517, 48042, 28695, -32163, -2518, -78754, 30260, -84215, -72182, 74194, 92783, 68010, 47310, 16031, 45175, -46068, 98174, -99871, 74934, -45336, -54311, 79278, 24674, -40377, -39153, -79384, 5765, 61928, -54214, 63888, -95233, -52327, -84372, 53599, 84395, 95139, -40414, -27202, -10650, 7066, -96753, 70946, 52286, 17815, 17961, -65257, -95478, -96068, -67841, 4010, -2798, -79330, 43150, -56464, -18730, -29734, 55827, 33629, 23532, -36364, -35832, -45169, -59444, -8017, 92593, -25108, 33271, -48892, -27286, 87771, -11440, -32440, -72704, -30888, 10141, -15697, -89283, 87354, -28129, -65860, -33026, 64838, 42114, -86588, -60198, 2213, -82435, -18378, -33925, -70542, 83948, -22575, 87628, 80251, 12220, -17822, -73205, 8392, 50187, -27642, -7951, -25403, 34795, -50716, -13090, -43554, 84927, 94321, 83848, 72541, 22002, 37016, 37754, -62058, 824, -10730, -80813, 37261, 75716, -50041, -6344, -66148, 26739, -25483, -49086, 52731, 35609, 88683, -26079, -89477, 38402, -63838, 94856, 23460, -92759, 18843, 61071, -89901, 6057, -77048, -79237, -3543, 4307, 64392, 60424, -46704, -63207, -50568, -30535, -26344, 47227, -32414, -11203, -53836, 42546, 32803, -27631, -63831, -80947, 76683, 67577, -87463, 33373, -60600, 75696, 88775, 60044, -5, -74085, -72030, -14294, -1018, 82894, 8118, 99426, -78343, -86107, -18863, 46821, -31224, 84481, 40645, 84044, 96621, -72864, -95416, 29272, 27313, 23988, 79279, -12158, -94069, -33142, 10085, 52552, -70006, 79754, 75021, 56808, -73718, -7022, 18729, 96463, 97629, -48496, -35247, 79575, 65009, 47439, -17534, -70242, -44060, -57411, -52364, -83341, -81033, -55764, 5461, 48469, -93123, 58926, -25507, 35517, 37848, 64615, 82225, -68002, 4512, 21034, -46802, 51318, 29734, -36977, -21670, -37226, 75351, 90197, -20483, 85907, -54602, 13972, 86970, -1353, 87078, -66637, -19441, -76303, -10498, -69791, -34821, -80780, 99651, 26405, -57275, -43769, -86143, -42678, -40839, -4136, -91982, -47208, -27818, -66215, -82669, -43544, -20067, -33203, -59051, 96846, -26115, 54034, -80100, -76476, -37456, 47967, 42408, 37441, -82189, -48928, 19117, 77453, -49784, -77690, 41137, -96384, -92813, -10554, 88293, -76388, -94823, 98062, -76365, 59625, 64291, -59815, 99467, -10994, 99481, -35696, -80903, -37247, -51175, 34650, -21424, -56427, 47422, -29043, -32767, 72388, 64657, -34149, -8764, 13689, -82413, -22788, 54567, -81194, -29385, -23062, 21257, 31963, -83142, 57382, -99527, -54668, 36191, -78657, 33828, 96222, -12481, -73863, -97812, 89464, -23505, -28846, -85951, 2518, -59028, -96611, 8279, -24102, -273, -976, -8689, 25022, -19064, -8648, 89771, -19642, 90409, 34075, -11917, 84908, 41438, 12212, -87471, 41841, 72447, -86499, 5190, 28801, -51280, 21041, 41071, 86093, -83690, -84738, 37868, -23301, 46057, -65247, 58631, 76761, 87316, -83463, 48358, -74223, 2517, 49009, 31076, -67049, -88640, 26170, 11749, 74577, 58607, 80309, 27515, -59517, -93401, -39774, 23125, 49283, 15532, -21189, 99923, 67261, -14140, -2679, 83882, 20855, -58333, 30416, -68924, 78716, 28245, 49258, -40401, 30538, -26076, -25855, -33596, 63243, -79797, 54191, -83051, -61280, -25462, 19227, -10064, -74409, 63001, 3266, -48661, -28758, -72269, -79775, 96835, 67983, 41570, -15900, 70402, 35248, -39349, -81371, -75136, 74480, 16006, -48889, 24200, -19850, 28808, 86096, 52651, -80319, 50075, 76658, 29568, 12940, -30181, 71774, 93700, 14296, -67918, 75305, -58685, -12931, -43220, 8728, -79656, -5996, 3860, -15596, -12548, 72470, -65283, -27983, -97160, 70710, -11926, 66679, 97630, 50737, 57165, 47982, 53982, -9808, -865, -21756, -22903, -33631, -50981, 11134, 25639, 50069, -41666, 71631, 15839, -36481, -84788, -8760, 16527, -13755, -59215, -39923, -78523, -18231, 7564, 91030, 60955, 46372, 62180, 75980, 54297, 81670, 99848, -88301, 82009, 88897, -31566, -89748, 6105, 76234, 89495, -25465, 13230, 39838, 81955, 58443, 90241, 13490, -45208, -18706, -41503, 23678, -73603, -9806, 79612, -33872, 52626, 92450, 66617, -6509, 85629, -56283, -50689, 22969, -6081, 76280, -22439, 66099, 96521, -14983, -17963, 13510, -23879, -37927, 22792, 49898, -57710, -65937, 5656, 35871, 43457, -60750, -1063, -74905, -28043, 3115, -35049, 38557, 2206, 47101, -41286, -71511, 33858, -30442, -64289, 38802, 8322, -33207, -25182, 24588, 54850, 15394, 62054, -98311, -47790, -25632, 28526, -45651, 91035, 81197, -98449, 73822, -93790, -79322, -11544, 69875, -11037, -33827, -89480, 704, 52563, -78350, -33917, 35687, 61402, -26201, -78130, 26630, 42992, 83007, 58884, -31725, -90404, 40385, -40338, 1639, -58922, 93215, 31499, -48004, 96980, -8798, -77298, -53439, -96460, -42202, -65221, 28633, -33651, 64143, -68777, -76209, -8445, 39845, -69279, 27230, 69929, 99880, 14407, -50731, 28450, -15490, 53826, -83852, 76790, -50473, 42111, -75401, 40025, 33215, 13565, -28345, -61708, 33301, 27942, -35894, -50726, -9682, 87974, -1911, -95129, 38920, -27084, -83123, 8272, -28026, -43818, 63644, 91152, 60520, -46561, -48174, 10708, 13411, 66202, -65731, 23020, -85990, -81487, 51347, -6340, 2892, 33982, -3287, 1260, -18098, -83666, 60746, -70753, -28744, 56442, -88964, -16516, -82406, 12963, -38946, -48100, 16718, 17946, 76430, -235, -2358, -31931, 73633, -28587, -56260, 45035, 5698, -60355, 32099, 17534, -88052, -33053, 20109, -64184, -197, 63935, -41247, 54268, 59851, 92216, 92941, 91652, -47051, 4306, -24652, 79904, -32074, 7957, -76413, -82387, 92267, -23165, 97550, -71518, -60700, 39451, 57213, 70071, -97081, -55995, 32256, -62130, -24016, -41741, -17504, -15479, -95256, 86781, 4364, 60057, -97316, -8004, 8579, -58006, 14291, 94410, 82122, 31673]
A_s = insertion_sort(A)
print(f"OG: {A}")
print("---")
print(f"Sorted: {A_s[0]}")
print(f"Swaps: {A_s[1]}")
```
